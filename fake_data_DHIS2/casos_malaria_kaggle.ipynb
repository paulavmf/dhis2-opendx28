```python
def conver_to_time_series(df):
    # Assuming you have a DataFrame named 'df' with columns: 'Date', 'Variable1', 'Variable2'
    df['Year'] = pd.to_datetime(df['Year'], format = "%Y")
    # .dt.strftime("%Y")
    df.set_index('Year', inplace=True)

    # Optional: Sort the DataFrame by date
    df.sort_index(inplace=True)
    return df
```

```python
import pandas as pd
import numpy as np

# Crear DataFrame con los datos proporcionados
data = {'Year': [2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017],
        'AggregatedValue': [118332.0, 241926.0, 165933.0, 330331.0, 274119.0, 280241.0,
                            366687.0, 268912.0, 492253.0, 349540.0, 395706.0]}

def convert_to_weekly_normal(data):
    df = pd.DataFrame(data)

    # Establecer el año como índice
    df.set_index('Year', inplace=True)

    # Crear DataFrame vacío para los datos semanales
    weekly_data = pd.DataFrame()

    # Iterar por cada año
    for year, row in df.iterrows():
        total_year_value = row['AggregatedValue']
        weekly_values = np.random.normal(loc=total_year_value / 52, scale=total_year_value / 52 / 4, size=52)  # Generar valores semanales
        weekly_data = weekly_data.append(pd.DataFrame({'Year': year, 'Value': weekly_values}))

    # Restablecer el índice del DataFrame resultante
    weekly_data.reset_index(drop=True, inplace=True)

    # Convertir los valores a enteros
    weekly_data['Value'] = weekly_data['Value'].astype(int)
    weekly_data.to_csv('extrapolated_data.csv')
    # Imprimir el resultado
    return weekly_data
```

```python
import pandas as pd
import statsmodels.api as sm


def weekly_interpolate_decompose(yearly_data):
# Assuming you have a DataFrame named 'yearly_data' with columns: 'Year' and 'AggregatedValue'

    # Step 1: Prepare the DataFrame with yearly aggregated data

    # Step 2: Decompose the time series
    res = sm.tsa.seasonal_decompose(yearly_data['AggregatedValue'], model='additive', period = 4)
    trend = res.trend
    seasonal = res.seasonal
    residual = res.resid

    # Step 3: Model the components
    trend_model = sm.tsa.ARIMA(trend.dropna(), order=(1, 0, 0))
    trend_fit = trend_model.fit()

    seasonal_model = sm.tsa.ExponentialSmoothing(seasonal, seasonal_periods=12, trend=None, seasonal='add')
    seasonal_fit = seasonal_model.fit()

    # Step 4: Extrapolate the components
    extrapolated_trend = trend_fit.predict(start=len(trend), end=len(trend) + 51)  # Extrapolate for 52 weeks (1 year)
    extrapolated_seasonal = seasonal_fit.predict(start=len(seasonal), end=len(seasonal) + 51)

    # Step 5: Recompose the time series
    extrapolated_values = extrapolated_trend + extrapolated_seasonal + residual.mean()

    # Step 6: Add random variation (optional)
    random_noise = residual.mean() * pd.Series(data=np.random.normal(0, 0.2, len(extrapolated_values)), index=extrapolated_values.index)
    extrapolated_values += random_noise

    # Create a DataFrame for the extrapolated time series
    weekly_data = pd.DataFrame({'Week': pd.date_range(start='2022-01-01', periods=len(extrapolated_values), freq='W'), 'Value': extrapolated_values})

    # Optional: Plot the extrapolated data to visualize the results
    weekly_data.plot(x='Week', y='Value')

    # Optional: Save the extrapolated data to a CSV file
    weekly_data.to_csv('extrapolated_data.csv', index=False)
    return weekly_data
```
