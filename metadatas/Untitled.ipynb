```python
with open('../metadatas/opendx_orgUnits_no_geometries.json') as f:
    # Load JSON data from file
    data = json.load(f)
```

```python
data
```

```python
URL = "http://localhost:8080"
```

```python
try:
    url = URL + "/api/me"
    session = requests.Session()
    session.auth = ('admin', 'district')
    url = f"{URL}/api/organisationUnits"
    response = session.post(url, json=json.dumps(data), headers={'content-type': 'application/json'})
    print("[CREATE ORGANISATION UNIT] >> "+ response.text)
except Exception:
    print("Unspected exception")
```
